```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/kuntik/carries_dataset/annotations.csv')
```

```python
n_empty = 0
n_caries = 0
width = []
height = []

for index, row in df.iterrows():
    if row.noobject:
        n_empty += 1
        continue
    else:
        n_caries += 1
        width.append(row.xmax - row.xmin)
        height.append(row.ymax - row.ymin)
```

```python
n_caries
print(len(df.image.unique()))
print(n_empty)
print(len(df.image.unique())- n_empty)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax =plt.subplots(figsize=(15,12))
ax.set(xscale='log', yscale='log')
# , xlabel='box width', ylabel='box height', fontsize = 30)
plt.title('Distribution of box sizes', fontsize= 30)
f.set_dpi(300)
plt.grid()
ax.tick_params(axis='both', which='major', labelsize=15)
ax.tick_params(axis='both', which='minor', labelsize=15)
scatter = sns.scatterplot(x=width, y=height, ax=ax)
scatter.set_ylabel('box_height [px]', fontsize = 20)
scatter.set_xlabel('box_width [px]', fontsize = 20)
plt.savefig('boxes_scatter.png')
```

```python
np_height = np.asarray(height)
np_width = np.asarray(width)
print(f"width min : {min(width)}, width max : {max(width)}, height min : {min(height)}, height max : {max(height)}")
print(f"width min : {np_width.mean()}, width max : {np_width.std()}, height min : {np_height.mean()}, height max : {np_height.std()}")
```
